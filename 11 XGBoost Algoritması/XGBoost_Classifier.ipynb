```python
dataset = pd.read_csv("Churn_Modelling.csv")
dataset
```

```python
dataset = dataset.drop(["RowNumber","CustomerId","Surname","Geography"],axis=1)
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoder.fit(dataset["Gender"].values)
gender_encoded = encoder.transform(dataset["Gender"].values)
dataset["Gender"] = gender_encoded  # Female->0 , Male->1
dataset
```

```python
Y = dataset["Exited"].values  # müşterinin dönüşü olumlu (1) veya olumsuz (0)
X = dataset.drop("Exited",axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.33)
```

```python
from xgboost import XGBClassifier
model = XGBClassifier()             #parametreler ayarlanabilir
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
result = confusion_matrix(y_pred,y_test)
result
```
