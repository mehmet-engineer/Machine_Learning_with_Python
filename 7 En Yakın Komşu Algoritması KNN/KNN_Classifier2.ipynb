```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Iris.csv")
data
```

```python
data = data.drop("Id",axis=1)
data
```

```python
# 3 label classes --> Iris-setosa = 0, Iris-versicolor = 1, Iris-virginica = 2
```

```python
Y = data["Species"].values
data = data.drop("Species",axis=1)
```

```python
from sklearn import preprocessing
encoder = preprocessing.LabelEncoder()
encoder.fit(Y)
Y_numbers = encoder.transform(Y)
Y_numbers
```

```python
Y_num_serisi = pd.Series(Y_numbers, name="Class")
dataframe = pd.concat([data,Y_num_serisi],axis=1)
dataframe
```

```python
Y = dataframe["Class"].values
X_1234 = dataframe.drop("Class",axis=1)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_1234,Y,test_size=0.33)
```

```python
from sklearn.neighbors import KNeighborsClassifier
#n_neighbors = int(sqrt(len(x_train))/2)
knn = KNeighborsClassifier(n_neighbors=5,metric="euclidean")
knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)
```

```python
y_pred_seri = pd.Series(y_pred)
y_test_seri = pd.Series(y_test)
a = dict(Gercek_Sinif=y_test_seri, Tahmin_Sinif=y_pred_seri)
dataFrame = pd.DataFrame(a)
dataFrame
```

```python
from sklearn.metrics import confusion_matrix
result = confusion_matrix(y_test,y_pred)
result
```
