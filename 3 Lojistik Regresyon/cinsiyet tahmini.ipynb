```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("veriler.csv")
Y = data["cinsiyet"].values
X1 = data["kilo"].values
X2 = data["yas"].values
X3 = data["boy"].values
data
```

```python
X_123 = data.drop("cinsiyet",axis=1)
X_123
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_123,Y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(x_train,y_train)
tahmin = log_reg.predict(x_test)
tahmin
```

```python
y_test
```

```python
tahminler = pd.Series(tahmin)
gercek_degerler = pd.Series(y_test)
data = dict(cinsiyet=gercek_degerler, tahmin=tahminler)
comparison_df = pd.DataFrame(data)
comparison_df
```

```python
# sınıflandırma değerlendirmesi CONFUSION MATRIX
from sklearn.metrics import confusion_matrix
result = confusion_matrix(y_test,tahmin)
result
```
